# 1) What were the ages of the workers when they left the company?

```python
plt.hist(attrition['Age']) 
plt.title('Age And Attrition')
plt.xlabel('Ages')
plt.ylabel('Attrition')
plt.show()
```

The histogram above depicts the ages of employees and the numbers of employees leaving the IBM company due to attrition. The x-axis covers an age range from zero to sixty while the y-axis covers attrition numbers from zero to two hundred and fifty. The histogram indicates that approximately two-thirds of the IBM workers left the company when they were in their late twenties to their early forties. It also indicates that approximately a tenth of the employees left the company when they were in their late tens to their mid-twenties and that approximately a fifth of the workers left the company when they were in their mid-forties to their late fifties.

```python
plt.boxplot(attrition['Age']) 
plt.title('Age And Attrition')
plt.xlabel('Attrition')
plt.ylabel('Ages')
plt.show()
```

The box plot above depicts the attrition of the IBM employees and the ages of the employees when they left IBM just like the histogram does. The x-axis covers the attrition and the y-axis covers the age range of the employees from twenty to sixty. The first quartile group consists of workers who left IBM when they were in their late tens to their late twenties while the second quartile consists of workers who left the company when they were in their early thirties to their mid-thirties. The third quartile group consists of employees who left IBM when they were in their late thirties to their early forties while the fourth quartile consists of employees who left IBM when they were in their mid-forties to their late fifties. The minimum of the boxplot is approximately at age eighteen, the maximum is approximately at age sixty, the lower quartile is at about age thirty, the upper quartile is at about age forty-four, and the median is at about age thirty-six. The median of the box plot is thirty-six years of age and this median corresponds with the age range from the late twenties to the early forties from the previous histogram. 

I hypothesize that the median is thirty-six and that the age range is from the late twenties to the early forties because of educational levels, work-life balance, and job levels. The older employees are more likely to have higher education levels than the younger employees and the education levels of the younger employees might have contributed to their job levels being lower than their older associates, so this would mean that the younger workers might have received lower pay and have less power in the company thereby contributing to their departure from IBM. Also, the older workers most likely are better at maintaining a work-life balance, so that would explain why the younger, more inexperienced workers would leave IBM. The younger workers having less of a work-life balance than their older associates would mean that the younger workers who have less of a chance to do their jobs at top capacity. In addition, the older employees would most likely have higher job levels than their younger associates because they were with the company longer, so this would also mean that the younger employees would receive lower pay thereby contributing to their departure from the company. Therefore, education, work-life balance, and job levels are three important factors for why the younger employees might have left the company.

The first group consists of departed employees who had relationship satisfaction levels from one to two. The second group consists of departed workers who had relationship satisfaction levels from two to three and the third group consists of departed workers who had relationship satisfaction levels from three to four. The minimum is at a relationship satisfaction level of one, the lower quartile is at a level of two, the upper quartile is at a level of four, and the median is at a level of three. The median of the box plot is three and this median corresponds with the relationship satisfaction levels of three and four from the previous histogram. 

I hypothesize that the median is three and that the relationship satisfaction levels with the highest attrition are three and four because of marital status, work-life balance, and performance rating. I suspect that the younger employees have more marriages and fewer single people and divorced people while the older employees have fewer marriages and more single people and divorced people. This means that the younger employees might be more occupied with their love lives, which would contribute to their departure from the IBM company. I also suspect that the younger employees would have a weaker balance between their personal lives and their work lives because they would be more occupied with their mates. In addition, the performance ratings of the younger workers would be weaker than the performance ratings of the older workers because the older workers are more focused on their jobs. Therefore, marital status, work-life balance, and performance rating are three significant factors for why workers with higher relationship satisfaction levels would leave IBM.

# 3) Were workers who had the highest amount of working years more likely to leave the company?

The histogram depicts the total working years of the workers and the attrition numbers of the workers. Approximately three-fourths of the employees who left IBM worked from zero to sixteen years before they left IBM. In addition, approximately a fourth of the employees worked from sixteen to forty years. 

```python
plt.boxplot(attrition['TotalWorkingYears']) 
plt.title('Total Working Years And Attrition')
plt.xlabel('Attrition')
plt.ylabel('Total Working Years')
plt.show()
```

The box plot depicts the total working years of the IBM employees and the attrition numbers of the employees just as

second quartile consists of six to ten working years. The third quartile consists of ten to fifteen years while the fourth quartile consists of fifteen to twenty-eight working years and there are eleven outliers beyond the maximum. The minimum is at zero total working years, the lower quartile is at six total working years, the upper quartile is at fifteen total working years, the maximum is at twenty-eight total working years, and the median is at ten total working years.

I hypothesize that the median is ten and that the span of total working years with the highest attrition is zero to sixteen years because of the years the employees worked at IBM, the percent salary hike, the number of companies worked at, and the stock option level. I suspect that the years the employees worked at IBM contributed to the departure of the workers because some of the younger employees are not as dedicated to their jobs as their older peers. I suspect that the workers who have fewer total working years left IBM because they had lower percent salary hikes. Also, I suspect that the employees left IBM because they worked for few companies; the issue is that they were inexperienced with working under different managements, which means that they were somewhat ill equipped for their jobs at IBM. Another reason the workers might have left IBM is the stock option levels of IBM; the older employees most likely had higher stock option levels than the younger employees had, so the younger employees left because they had very little influence via stock in the company. Therefore, the years the employees worked at IBM, the percent salary hikes, the numbers of companies worked at, and the stock option levels were four significant factors for why workers with fewer total working years would leave IBM.
