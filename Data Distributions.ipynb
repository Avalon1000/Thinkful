```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
variable1 = np.random.gamma(100, 100, 100)

df = pd.DataFrame({'rand1': variable1})

df.hist()

mean = np.mean(variable1)
sd = np.std(variable1)

print(mean)
print(sd)

plt.axvline(x=mean,color='green')
plt.axvline(x=mean+sd, color='purple')
plt.axvline(x=mean-sd, color='purple')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

variable1 = np.random.normal(5, 0.5, 100)
variable2 = np.random.normal(10, 1, 100)
variable3 = variable1 + variable2

mean = np.mean(variable3)
sd = np.std(variable3)

print(mean)
print(sd)

plt.hist(variable3)
plt.axvline(x=mean, color='green')
plt.axvline(x=mean+sd, color='purple')
plt.axvline(x=mean-sd, color='purple')
plt.show()
print('The results will vary depending on the random seed. Also, the statistics display some skew, in general.')
```
