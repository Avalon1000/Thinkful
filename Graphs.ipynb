```python
import pandas as pd
import matplotlib.pyplot as plt

titan = pd.read_excel(r'C:\Users\Darin Peacock\Documents\Titanic2.xlsx')
titan.head()
```

```python
plt.plot(titan['Age'])
plt.title('Ages of Passengers')
plt.xlabel('Passengers')
plt.ylabel('Ages')
plt.show()

plt.plot(titan['Fare'])
plt.title('Fares of Passengers')
plt.xlabel('Passengers')
plt.ylabel('Fares')
plt.show()

plt.plot(titan['SibSp'])
plt.title('SibSps of Passengers')
plt.xlabel('Passengers')
plt.ylabel('SibSps')
plt.show()
```

```python
plt.hist(titan['Age']) 
plt.title('Ages of Passengers')
plt.xlabel('Ages')
plt.ylabel('Passengers')
plt.show()
```

```python
plt.boxplot(titan['Pclass'])
plt.title('Pclasses of Passengers')
plt.xlabel('Passengers')
plt.ylabel('Pclasses')
plt.title('Pclass Box Plot')
plt.show()
```

```python
plt.boxplot(titan['SibSp'])
plt.title('SibSp Box Plot')
plt.xlabel('Passengers')
plt.ylabel('SibSps')
plt.show()
```
