```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

titanic = sns.load_dataset("titanic")
print(titanic.head(n=4))
```

```python
# 1A) One variable plotted in four different ways
```

```python
plt.hist(titanic['age'])
plt.xlabel('Ages', fontsize=14)
plt.ylabel('Passengers', fontsize=14)
plt.title('Ages And Passengers', fontsize=14)
plt.show()

g = sns.distplot(titanic['age'], hist=False, bins=10)
plt.xlabel('Ages', fontsize=14)
plt.ylabel('Population Density', fontsize=14)
plt.title('Ages And Density', fontsize=14)
plt.show()

g = sns.set_style("darkgrid")
plt.plot(titanic['age'])
plt.xlabel('Passengers', fontsize=14)
plt.ylabel('Ages', fontsize=14)
plt.title('Ages And Passengers', fontsize=14)
plt.show()

plt.style.use('seaborn')
titanic['age'].plot.area()
plt.xlabel('Passengers', fontsize=14)
plt.ylabel('Ages', fontsize=14)
plt.title('Ages And Passengers', fontsize=14)
plt.show()
```

```python
# The plots here depict the age ranges of the passengers on the Titanic. The first plot is a histogram, the second plot is a 
# Kernel density estimation plot, the third plot is a line plot, and the fourth plot is a stacked area chart. The histogram 
# indicates that the majority of the passengers are between about sixteen and forty years of age. The Kernel density estimation 
# plot indicates that the population density of the passengers is the highest between about ten and fifty years of age. The 
# line plot indicates that the majority of the passengers are between about eighteen and about forty-two years of age. The 
# stacked area chart indicates that the vast majority of the passengers are between zero to about forty years of age. 
# Therefore, the histogram, the Kernel density estimation plot, the line plot, and the stacked area chart are accurate. 
```

```python
# 1B) Two continuous variables plotted in three different ways
```

```python
titanic = sns.load_dataset("titanic")

g = sns.jointplot(x=titanic['age'], y=titanic['fare'], data=titanic)
plt.show()

g = sns.jointplot(x=titanic['age'], y=titanic['fare'], kind="hex", color="k")
plt.show()

g = sns.jointplot(x=titanic['age'], y=titanic['fare'], data=titanic, kind="kde")
plt.show()
```

```python
# The plots here depict the relationship between the ages of the passengers on the Titanic and the fares that these passengers 
# have to pay. The first plot is a scatterplot, the second plot is a hexbin plot, and the third plot is a Kernel desnity 
# estimation. The scatterplot indicates that the majority of the passengers are zero to sixty years old and that most of them 
# have to pay fares that are under one hundred dollars. The hexbin plot indicates that the vast majority of the passengers are 
# between the ages of about eighteen to about thirty-two years of age and pay around ten dollars for fares. The Kernel 
# density estimation plot indicates that the vast majority of the passengers are about sixteen to thirty-four years old and 
# that this group of passengers pays about ten dollars for fares.  
```

```python
# 1C) One continuous variable and one categorical variable plotted in six different ways
```

```python
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib as mpl
import seaborn as sns
%matplotlib inline
titanic = sns.load_dataset("titanic")

sns.catplot(x="sex", y="age", data=titanic)
plt.xlabel('Sex', fontsize=14)
plt.ylabel('Age', fontsize=14)
plt.title('Ages And Sexes')
plt.show()

sns.catplot(x="age", y="sex", kind="violin", data=titanic)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Sex', fontsize=14)
plt.title('Ages And Sexes')
plt.show()

sns.catplot(x="sex", y="age", kind="box", data=titanic)
plt.xlabel('Sex', fontsize=14)
plt.ylabel('Age', fontsize=14)
plt.title('Ages And Sexes', fontsize=14)
plt.show()

sns.catplot(x="sex", y="age", kind="boxen", data=titanic.sort_values("age"))
plt.xlabel('Sex', fontsize=14)
plt.ylabel('Age', fontsize=14)
plt.title('Ages And Sexes')
plt.show()
```

```python
sns.catplot(x="sex", y="survived", hue="class", kind="bar", data=titanic)
plt.xlabel('Sex', fontsize=14)
plt.ylabel('Survived', fontsize=14)
plt.title('Sexes And Survivors', fontsize=14)
plt.show()

sns.catplot(x="sex", y="survived", hue="class", kind="point", data=titanic)
plt.xlabel('Sex', fontsize=14)
plt.ylabel('Survived', fontsize=14)
plt.title('Ages And Survivors', fontsize=14)
plt.show()
```

```python
# Four of the plots here depict the relationship between the sexes and ages of the passengers on the Titanic while two of the 
# plots depict the relationship between the sexes of the passengers and the number of passengers who survived the sinking of 
# the Titanic. The categorical scatterplot indicates that the majority of the passengers are male and that there are more 
# older males than older females. The violinplot indicates that the highest density for males is between about twelve and about 
# forty-two years of age and that the highest density for females is between about twelve and about forty-two years of age. 
# The boxplot indicates that most of the males are between about twenty-two and about forty years of age (with four upper 
# outliers) and that most of the females are between about nineteen and thirty-nine years of age. The boxenplot indicates that 
# most of the males are between twenty-two and forty years of age (with one upper outlier) and that most of the females are 
# between nineteen and thirty-nine years of age. The bar plot indicates that the males have survival levels of about 0.39 
# (first class), 0.18 (second class), and 0.17 (third class) while females have survival levels of about 0.98 (first class), 
# 0.91 (second class), and 0.5 (third class). The point plot indicates that the males have survival levels of 0.39 (first 
# class), 0.18 (second class), and 0.17 (third class) and that females have survival levels of 0.98 (first class), 0.91 (second 
# class), and 0.5 (third class). The findings of the four plots regarding sex and age are accurate because they are all similar 
# and the findings of the other two plots for survivors and sex are accurate because they are the same.     
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
mpg = sns.load_dataset("mpg")
print(mpg.head(n=20))
```

```python
# 2A) Probability distributions
```

```python
g = sns.PairGrid(mpg)
g.map_diag(sns.kdeplot)
g.map_offdiag(sns.kdeplot, n_levels=6)
```

```python
# The master plot here is composed of Kernel density estimation plots and these plots vary across the master plot. The plots 
# for mpg vs. mpg, cylinders vs. cylinders, displacement vs. displacement, horsepower vs. horsepower, weight vs. weight, 
# acceleration vs. acceleration, model_year vs. model_year have only a single line that moves different ways per plot. The plots 
# for the two model_year sets (column and row) have bulky densities and the plots for the two acceleration sets have 
# rough-looking densities. The plots for the two weight sets have curved densities (with one of them being a little scattered) 
# and the plots for the two horsepower sets are nonexistant save for one line for where the two horsepower graphs meet. Also, 
# two of the plot for the displacement sets are curved while two more of them are scattered, the two plots for the cylinders 
# are unified, and the plots for the mpg sets are unified and curved. These are the formations of the master plot for Kernel 
# density estimation plots.      
```
