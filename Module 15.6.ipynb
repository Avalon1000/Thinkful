```python
wellcome = pd.read_excel(r'F:\Downloads\WELLCOME_APCspend2013_forThinkful.xlsx')
wellcome.head()
```

```python
wellcome['Journal title'] = wellcome['Journal title'].replace(['PLOSONE', 'PLOS ONE', 'PLOS 1', 'PLOS'], 'PLOS ONE')

wellcome['Journal title'] = wellcome['Journal title'].replace(['ACTA D', 'ACTA CRYSTALLOGRAPHICA SECTION D', 
'ACTA CRYSTALLOGRAPHICA D', 'ACTA CRYSTALLOGRAPHICA, SECTION D', 'ACTA CRYSTALLOGRAPHICA SECTION D, BIOLOGICAL CRYSTALLOGRAPHY'], 
'ACTA CRYSTALLOGRAPHICA SECTION D: BIOLOGICAL CRYSTALLOGRAPHY')

wellcome['Journal title'] = wellcome['Journal title'].replace(['AMERICAN JNL EPIDEMIOLOGY'], 'AMERICAN JOURNAL OF EPIDEMIOLOGY')

wellcome['Journal title'] = wellcome['Journal title'].replace(['ANTIMICROBIAL AGENTS AND CHEMOTHERAPY', 
'ANTIMICROBIAL AGENTS AND CHEMOTHERAPY'], 'ANTIMICROBIAL AGENTS & CHEMOTHERAPY')

wellcome['Journal title'] = wellcome['Journal title'].replace(['ANGEWANDTE CHEMIE', 'ANGEWANDTE CHEMIE INTERNATIONAL EDITION', 
'ANGEWANDTE INT ED'], 'ANGEWANDTE CHEMIE')

wellcome['Journal title'] = wellcome['Journal title'].replace(['BEHAVIOUR RESEARCH AND THERAPY'], 'BEHAVIOUR RESEARCH & THERAPY')

wellcome['Journal title'] = wellcome['Journal title'].replace(['BIOCHEM JOURNAL', 'BIOCHEMICAL JOURNALS'], 'BIOCHEMICAL JOURNAL')

wellcome['Journal title'] = wellcome['Journal title'].replace(['BIOCHEM SOC TRANS'], 'BIOCHEMICAL SOCIETY TRANSACTIONS')

wellcome['Journal title'] = wellcome['Journal title'].replace(['BRITISH JOURNAL OF OPHTHALMOLOGY'], 'BRITISH JOURNAL OF OPHTHALMOLOGY')

wellcome['Journal title'] = wellcome['Journal title'].replace(['CELL DEATH DIFFERENTIATION'], 'CELL DEATH & DIFFERENTIATION')

wellcome['Journal title'] = wellcome['Journal title'].replace(['CHILD: CARE, HEALTH DEVELOPMENT'], 'CHILD: CARE, HEALTH & DEVELOPMENT')

wellcome['Journal title'] = wellcome['Journal title'].replace(['CURR BIOL'], 'CURRENT BIOLOGY')
```

```python
wellcome['Journal title'] = wellcome['Journal title'].apply(lambda x: str(x).strip().upper())
```

```python
wellcome[ 'Journal title' ] = wellcome[ 'Journal title' ].fillna('')

wellcome[ 'journal_cleaned' ] = wellcome[ 'Journal title' ].str.replace('\d+', '')

wellcome[ 'journal_cleaned' ] = wellcome[ 'journal_cleaned' ].str.replace('[^\w\s]','' )

wellcome[ 'journal_cleaned' ] = wellcome[ 'journal_cleaned' ].str.strip()

wellcome[ 'journal_cleaned' ] = wellcome[ 'journal_cleaned' ].apply(lambda x: ''.join([word for word in x.split()]))
                                                                
wellcome[ 'journal_cleaned' ] = wellcome[ 'journal_cleaned' ].str.upper()

print('The Top 5 Journals and total Articles for each:')
print('(with a tie for 5th)')
wellcome['journal_cleaned'].value_counts().head(6)
```
